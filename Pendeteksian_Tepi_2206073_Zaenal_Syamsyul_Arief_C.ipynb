# **Soal 1: Jenis-Jenis Tepi**

### 1. Empat Jenis Tepi
1. Tepi Curam (Step Edge)
- Definisi: Terjadi ketika terdapat perubahan intensitas yang mendadak antara dua area pada citra. Perubahan ini terjadi hampir secara vertikal, menciptakan transisi tajam antara area gelap dan terang.
- Karakteristik:
  - Profil intensitasnya seperti langkah (step), dengan perubahan yang instan.
  - Tipe tepi ini sering muncul pada objek dengan kontur yang jelas.
- Contoh: Tepi dinding hitam pada latar belakang putih.
2. Tepi Landai (Ramp Edge)
- Definisi: Perubahan intensitas terjadi secara bertahap, bukan mendadak. Transisi antar area lebih mulus dan memakan jarak lebih panjang.
- Karakteristik:
  - Profil intensitas memiliki kemiringan (ramp) tanpa perubahan tajam.
  - Biasanya muncul akibat efek pencahayaan atau gradien alami.
Contoh: Bayangan pohon yang memudar di atas permukaan tanah.
3. Tepi Garis (Line Edge)
- Definisi: Perubahan intensitas terjadi di tengah dua area dengan intensitas serupa. Area ini terlihat seperti garis tipis di antara latar belakang.
- Karakteristik:
  - Intensitas meningkat dan menurun di sekitar garis, membentuk profil berbentuk "gundukan" (bump).
  - Sering muncul pada fitur sempit seperti kabel atau celah kecil.
- Contoh: Kabel listrik hitam yang tergantung di langit terang.
4. Tepi Atap (Roof Edge)
- Definisi: Intensitas meningkat secara perlahan hingga mencapai puncak, lalu menurun secara perlahan, menyerupai atap rumah.
- Karakteristik:
  - Profil intensitas memiliki puncak yang lebar dan landai.
  - Biasanya muncul pada objek dengan permukaan melengkung atau area dengan gradien kecil.
- Contoh: Pantulan cahaya pada permukaan bola mengkilap.

### 2. Berikan contoh masing-masing jenis tepi dalam kehidupan sehari-hari
Contoh Jenis Tepi dalam Kehidupan Sehari-hari
1. Tepi Curam (Step Edge)
- Deskripsi: Tepi dengan perubahan intensitas yang tajam dan mendadak.
- Contoh Sehari-hari:
  - Tepi meja gelap yang terlihat jelas di atas lantai terang.
  - Pintu hitam di dinding putih.
  - Garis horizon antara laut biru dan langit cerah.
2. Tepi Landai (Ramp Edge)
- Deskripsi: Tepi dengan perubahan intensitas yang bertahap.
- Contoh Sehari-hari:
  - Bayangan pohon yang memudar di atas tanah.
  - Langit saat matahari terbenam, dengan gradasi warna dari kuning ke merah.
  - Dinding dengan efek pencahayaan lembut yang membuat transisi warna halus.
3. Tepi Garis (Line Edge)
- Deskripsi: Tepi yang muncul di antara dua area dengan intensitas serupa, membentuk garis tipis.
- Contoh Sehari-hari:
  - Kabel listrik hitam tergantung di langit yang cerah.
  - Garis batas trotoar dengan celah tipis di antara batu-batu penyusunnya.
  - Garis pelangi tipis di permukaan air setelah terkena minyak.
4. Tepi Atap (Roof Edge)
- Deskripsi: Tepi dengan perubahan intensitas yang mencapai puncak dan kembali menurun.
- Contoh Sehari-hari:
  - Pantulan cahaya matahari pada permukaan bola mengkilap.
  - Puncak gunung dengan gradien lembut akibat bayangan.
  - Sorotan lampu pada mobil yang berkilau, membentuk puncak cahaya lembut.

### 3. Mengapa tepi dengan derau memerlukan pendekatan khusus untuk pendeteksiannya?

1. Derau Menyembunyikan Tepi yang Sebenarnya
  - Derau menyebabkan perubahan intensitas yang acak dan tidak teratur, membuat tepi yang nyata sulit dikenali.
 - Contoh: Pada citra dengan tekstur kasar atau sinyal lemah, derau dapat menutupi tepi penting.
2. Kesalahan Deteksi (False Positives)
  - Algoritma pendeteksi tepi sering salah mengidentifikasi derau sebagai tepi karena pola intensitasnya mirip.
  - Hal ini menghasilkan deteksi tepi palsu yang membuat analisis citra tidak akurat.
3. Kehilangan Tepi (False Negatives)
 -  Derau juga dapat merusak struktur tepi yang sebenarnya, menyebabkan algoritma melewatkan tepi penting.
4. Perubahan Gradien yang Tidak Relevan
  - Tepi biasanya dideteksi dengan perubahan gradien intensitas. Derau menciptakan fluktuasi gradien kecil yang mengganggu proses deteksi.

### Solusi untuk Tepi dengan Derau
Untuk menangani masalah ini, digunakan pendekatan khusus seperti:

- Praproses dengan Penyaringan (Filtering):

  - Metode seperti Gaussian filter atau median filter digunakan untuk mengurangi derau sebelum pendeteksian.
  - Ini membantu mempertahankan tepi utama sambil menghilangkan perubahan intensitas acak.
- Algoritma yang Lebih Canggih:

  - Canny Edge Detection: Menggunakan smoothing dan thresholding adaptif untuk mendeteksi tepi dengan akurasi tinggi meskipun terdapat derau.
  - Laplacian of Gaussian (LoG): Menggabungkan peredaman derau dengan pendeteksian tepi berbasis perubahan gradien.
- Pemetaan Threshold:

  - Menggunakan nilai ambang tertentu untuk memisahkan tepi yang relevan dari noise.

### Pentingnya Pendekatan Khusus
Tanpa pengolahan derau yang baik, kualitas deteksi tepi akan menurun drastis, memengaruhi tugas analisis citra seperti segmentasi, pengenalan objek, dan rekonstruksi visual. Pendekatan khusus memastikan bahwa hanya tepi yang signifikan yang dipertahankan, sehingga menghasilkan data yang lebih bersih dan akurat.



# **Soal 2: Pendeteksian Tepi Menggunakan Operator Gradien**

### Operator Sobel

```python
from google.colab import files
uploaded = files.upload()  # Mengunggah file gambar
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca gambar

image_path = next(iter(uploaded.keys()))  # Mendapatkan nama file yang diunggah
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Sobel horizontal
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)

# Sobel vertikal
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)

# Magnitudo gradien
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

# Menampilkan gambar asli dan hasil Sobel
plt.figure(figsize=(10, 10))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Sobel X')
plt.imshow(np.abs(sobel_x), cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Sobel Combined')
plt.imshow(sobel_combined, cmap='gray')
plt.axis('off')

plt.show()
```

# **Soal 3: Perbandingan Operator Pendeteksi Tepi**

## **Perbandingan antara operator Sobel, Prewitt, dan Canny berdasarkan**
### Akurasi Pendeteksian
1. Sobel	:
  - Deteksi tepi cukup baik untuk perubahan intensitas besar (tepi yang jelas).
  - Sensitif terhadap derau karena tidak memiliki mekanisme peredaman derau bawaan.
2. Prewitt:
  - Mirip Sobel tetapi kurang akurat karena bobot kernelnya lebih sederhana.
  - Menggunakan kernel dengan nilai tetap (tidak memberikan bobot lebih besar pada piksel tengah).
3. Canny:
  - Sangat akurat karena memiliki langkah tambahan, seperti smoothing (Gaussian filter) dan thresholding.
  - Mampu mendeteksi tepi yang lemah tanpa menangkap terlalu banyak derau (false positives diminimalkan).

### Kesimpulan Akurasi:

- Canny adalah yang paling akurat, terutama untuk citra dengan tepi halus atau derau.
- Sobel memiliki akurasi lebih tinggi daripada Prewitt, tetapi keduanya lebih sederhana daripada Canny

### Kompleksitas Komputasi
1. Sobel:
  - Kompleksitas rendah karena hanya menghitung gradien menggunakan kernel 3×3.
2. Prewitt:
  - Kompleksitas lebih rendah dibandingkan Sobel karena kernel lebih sederhana dan tidak berbobot.
3. Canny:
Kompleksitas tinggi karena melibatkan:
  1. Penyaringan Gaussian untuk meredam derau.
  2. Perhitungan gradien (mirip Sobel).
  3. Non-maximum suppression untuk menipiskan tepi.
  4. Thresholding ganda untuk mengeliminasi false positives dan false negatives.

#### Kesimpulan Kompleksitas:

- Prewitt adalah yang paling sederhana dan efisien secara komputasi.
- Canny membutuhkan waktu lebih lama tetapi memberikan hasil yang jauh lebih akurat.
- Sobel berada di antara keduanya

## **Situasi di Mana Operator Canny Lebih Direkomendasikan**
Operator Canny lebih direkomendasikan pada situasi berikut:

1. Ketika Akurasi Sangat Penting:
  - Jika diperlukan deteksi tepi yang halus, akurat, dan tidak terpengaruh oleh derau.
  - Contoh: Aplikasi medis seperti segmentasi citra MRI atau CT scan.

2. Ketika Citra Mengandung Derau:
  - Karena Canny memiliki langkah smoothing (Gaussian filter), ini membantu mengurangi pengaruh derau sebelum pendeteksian tepi.
  - Contoh: Citra yang diambil dalam kondisi pencahayaan rendah atau menggunakan kamera dengan noise tinggi.

3. Ketika False Positives Perlu Diminimalkan:
  - Dengan thresholding ganda, Canny memastikan bahwa hanya tepi yang signifikan yang terdeteksi.
  - Contoh: Sistem pengenalan objek, seperti pada kendaraan otonom atau pengawasan keamanan.

4. Untuk Tepi Halus atau Kompleks:
  - Canny efektif pada citra dengan detail kompleks atau transisi intensitas bertahap.
  - Contoh: Pendeteksian kontur wajah dalam gambar atau video.

### Kesimpulan Umum
- Sobel
  - Akurasi Pendeteksian : Moderat
  - Kompleksitas Komputasi : Moderat
  - Situasi Penggunaan : Tepi kasar/tepi jelas

- Prewitt
  - Akurasi Pendeteksian : Rendah
  - Kompleksitas Komputasi : Rendah
  - Situasi Penggunaan : Deteksi cepat dan sederhana
- Canny
  - Akurasi Pendeteksian : Tinggi
  - Kompleksitas Komputasi : inggi
  - Situasi Penggunaan : Deteksi presisi dan bebas derau

Kesimpulanya :
- Gunakan Sobel untuk analisis cepat dengan akurasi sedang.
- Gunakan Prewitt jika efisiensi sangat penting dan kebutuhan akurasi rendah.
- Gunakan Canny untuk aplikasi yang membutuhkan akurasi tinggi atau menangani citra yang penuh derau.

# Soal 4: Pendeteksian Tepi pada Citra Noisy

```python
from google.colab import files
uploaded = files.upload()  # Mengunggah file gambar
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca citra asli
image_path = next(iter(uploaded.keys()))  # Mendapatkan nama file yang diunggah
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

# Menambahkan derau salt-and-pepper pada citra
def add_salt_pepper_noise(image, salt_prob, pepper_prob):
    noisy_image = image.copy()
    total_pixels = image.size

    # Menambahkan "salt"
    num_salt = int(total_pixels * salt_prob)
    coords = [np.random.randint(0, i - 1, num_salt) for i in image.shape]
    noisy_image[coords[0], coords[1]] = 255

    # Menambahkan "pepper"
    num_pepper = int(total_pixels * pepper_prob)
    coords = [np.random.randint(0, i - 1, num_pepper) for i in image.shape]
    noisy_image[coords[0], coords[1]] = 0

    return noisy_image

noisy_image = add_salt_pepper_noise(image, salt_prob=0.02, pepper_prob=0.02)

# Mengurangi derau dengan Gaussian filter
gaussian_filtered = cv2.GaussianBlur(noisy_image, (5, 5), 0)

# Deteksi tepi menggunakan Sobel
sobel_x = cv2.Sobel(gaussian_filtered, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(gaussian_filtered, cv2.CV_64F, 0, 1, ksize=3)
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

# Deteksi tepi menggunakan Canny
canny_edges = cv2.Canny(gaussian_filtered, 100, 200)

# Menampilkan hasil
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.title("Noisy Image (Salt-and-Pepper)")
plt.imshow(noisy_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.title("Gaussian Filtered Image")
plt.imshow(gaussian_filtered, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.title("Sobel Edge Detection")
plt.imshow(sobel_combined, cmap='gray')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.title("Canny Edge Detection")
plt.imshow(canny_edges, cmap='gray')
plt.axis('off')

plt.show()
```

    - Non-maximum suppression memastikan hanya tepi yang signifikan yang dideteksi, sehingga hasil terlihat lebih presisi.
  - Kelebihan:
    - Sangat baik dalam menangani derau dan menghasilkan hasil yang akurat.
    - Memberikan tepi yang lebih jelas, bahkan pada area dengan perubahan intensitas bertahap.
  - Kekurangan:
    - Lebih lambat dibandingkan Sobel karena kompleksitasnya lebih tinggi.
    - Perbandingan dengan Citra Asli
  - Sobel:
    - Mendeteksi tepi utama dengan baik tetapi kurang presisi karena masih terpengaruh oleh sisa derau.
    - Hasilnya lebih kasar dibandingkan dengan Canny.
  - Canny:
    - Memberikan hasil yang lebih mendekati tepi asli dalam citra, bahkan pada area dengan derau atau perubahan intensitas yang halus.

- Rekomendasi:
